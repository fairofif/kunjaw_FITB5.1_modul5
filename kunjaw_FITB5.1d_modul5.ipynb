```python
from plotly import __version__
import cufflinks as cf
import plotly.express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.set_config_file(sharing='public',theme='pearl',offline=True)
```

bebas pake library apapun buat di kelas aku, tapi output valuenya sama ya :) disini aku pake plotly biar diagramnya responsif kalo cursornya diarahin ke diagramnya bakal ada popup detail valuenya. Secara logika sama aja, cuma beda nama fungsinya dan variable2nya.

case 1

```python
dff = df3[df3['Gender'] == 'F']
dff['Total_Trans_Amt'].iplot(kind="hist", yTitle = 'Frequency', xTitle = 'Total Transaction',color='#68e8cc', bins=20)
```

case 2

```python
dfunder = df3[df3['Total_Trans_Amt'] < 2000]
dfmarital = dfunder['Marital_Status'].value_counts().rename_axis('Marital_Status').reset_index(name='counts')
print("legenda di sebelah kanan bisa diteken jadi ilang")
fig = px.pie(values=dfmarital['counts'], names=dfmarital['Marital_Status'])
fig.show()
```

case 3

```python
df3[df3["Card_Category"] == "Blue"].iplot(kind="scatter", mode="markers", x = "Months_on_book", y = "Avg_Utilization_Ratio", color='#68e8cc', xTitle='lama penggunaan kartu kredit dalam bulan', yTitle='rata-rata rasio penggunaan kartu kredit')
```

case 4

```python
dfstatus = df3['Attrition_Flag'].value_counts().rename_axis('Attrition_Flag').reset_index(name='counts')
dfstatus.iplot(kind="barh", y = "counts", x = "Attrition_Flag")
```

case 5

```python
dfincome = df3[df3['Income_Category'].isin(["Less than $40K","$40K - $60K","$60K - $80K"])]
dfplot = dfincome.groupby(["Attrition_Flag", "Income_Category"]).size().reset_index()
dfplot.rename(columns = {0:'counts'}, inplace=True)

fig = px.bar(dfplot, x="Attrition_Flag", y='counts', color='Income_Category')
fig.show()
```
